```python
import random
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}

class Card:

    def __init__(self,suit,rank):
        self.suit=suit
        self.rank=rank

    def __str__(self):
        return f"{self.rank} of {self.suit}"

    def val(self):
        self.val+=values[f"{self.rank}"]



class Deck:

    def __init__(self):
        self.deck = [ ]
        for suit in suits:
            for rank in ranks:
                ob=Card(suit,rank)
                self.deck.append(ob)

    def __str__(self):
            return f"{len(self.deck)}"

    def shuffle(self):
        random.shuffle(self.deck)

    def deal(self):
        return self.deck.pop()

class Hand:

    def __init__(self):
        self.cards = []
        self.value = 0
        self.aces = 0

    def add_card(self, card):

        self.cards.append(card)
        self.value+=values[f"{card.rank}"]
        if card.rank == 'Ace' and self.value>21 and self.value<=31:
            self.value -= 10

    def disp_hand(self):
        for i in range(len(self.cards)):
            print(self.cards[i])

    def disp_dealer_hand(self):
        print(self.cards[0])



    def val(self):
        return self.value


class Chips:


    def __init__(self,total):
        self.total=total
        self.bet = 0

    def take_bet(self):
        b = int(input("Enter the betting amount:"))
        self.bet = b
        if self.bet>self.total:
            while True:
                print("You cannot bet more than the total amount you have.")
                b = int(input("Enter the betting amount:"))
                self.bet = b
                if self.bet<=self.total:
                    break


    def win_bet(self):
        self.total+=self.bet
        print(f"You win {self.bet} rupees")

    def lose_bet(self):
        self.total-=self.bet
        print(f"You lose {self.bet} rupees")


def hit(deck, hand):
    deck.shuffle()
    hand.add_card(deck.deal())

t = int(input("Enter the total amount:"))
chips = Chips(t)
print(f"You have {chips.total} Rupees")




resp=input("Are you ready to play??(y/n:)")

while(resp=='y'):

    chips.take_bet()
    test_deck=Deck()
    test_deck.shuffle()
    phand=Hand()
    phand.add_card(test_deck.deal())
    phand.add_card(test_deck.deal())
    print("Your cards:")
    phand.disp_hand()
    print(phand.val())
    if phand.val()==21:
        print("You won!!you got 21.")
        chips.win_bet()
        print(f"Your available money is {chips.total}:")
    else:
        dhand=Hand()
        dhand.add_card(test_deck.deal())
        dhand.add_card(test_deck.deal())
        print()
        print("Dealer's hands:")
        dhand.disp_dealer_hand()
        ans='h'
        while(ans=='h'):
            ans=input("Do you want to hit or stay?(h/s):")
            if ans=='h':
                hit(test_deck,phand)
                phand.disp_hand()
                print(phand.val())
                if phand.val()==21:
                    print("You win!! you have got 21")
                    c=1
                    break
                if phand.val()>21:
                    print("Busted!! You lose. You went over 21")
                    c=-1
                    break
            else:
                break

        if phand.val()<21:
            print("Dealer's hands shown:")
            dhand.disp_hand()
            print(dhand.val())
            while(dhand.val()<17):
                hit(test_deck,dhand)
            print("After addition until 17..")
            print("Dealer's cards:")
            dhand.disp_hand()
            print(dhand.val())
            if dhand.val()>21:
                print("You win!! Dealer loses,he went over 21.")
                c=1
            elif dhand.val()==21:
                print("You lose!! Dealer got 21")
                c=-1

            elif dhand.val()>phand.val():
                print("Dealer wins!! He is closer to 21 than you.")
                c=-1
            elif phand.val()>dhand.val():
                print("You win!! you are closer to 21 than him.")
                c=1
            elif phand.val()==dhand.val():
                print("Tied match!!You neither win nor lose")
                c=0

        if c==1:
            chips.win_bet()
        elif c==-1:
            chips.lose_bet()
        elif c==0:
            pass

        print(f"Your available money is {chips.total}:")
        print()
        print()
        resp = input("Do you want to play again?(y/n)")
```
